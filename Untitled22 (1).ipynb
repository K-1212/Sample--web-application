- Dash :Dash is a python framework created by plotly for creating interactive web applications. 
- Dash apps are composed of two parts. The first part is the "layout" of the app and it describes what the application looks like. The second part describes the interactivity of the application
- The layout is composed of a tree of "components" such as html.Div and dcc.Graph.
- Div:A Div component. Div is a wrapper for the div HTML5 element
- graph :The dcc.Graph component can be used to render any plotly-powered data visualization, passed as the figure argument.
- callback:unctions that are automatically called by Dash whenever an input component's property changes, in order to update some property in another component (the output).
- dropdown:which is used to select the inputs to be shown at the output

```python
#updating the colors for the background and text
fig.update_layout(
    plot_bgcolor=colors['background'],
    paper_bgcolor=colors['background'],
    font_color=colors['text']
)
# creating the app
app=dash.Dash()
# creating the app .layout
app.layout = html.Div(children=[
    # All elements from the top of the page
    html.Div([
        html.H1(children='Sample Data',
                # adding style to the color ,text
               style={
                   'textAlign': 'center',
            'color': colors['text']
               }),
#name for the web application
        html.Div(children='''
            Dash: A web application framework for Python.
        ''',
                style={
                    'textAlign': 'center',
            'color': colors['text']
                   
                }),
# ploting the graph
        dcc.Graph(
            id='graph',
            figure=fig
        ),  
    ]),
                                  
    # New Div for all elements in the new 'row' of the page
    html.Div([ 
        dcc.Graph(id='first-graph'),
        html.Label([
            "colorscale",
            dcc.Dropdown(
                id='colorscale-dropdown', clearable=False,
                value='bluyl', options=[
                    {'label': c, 'value': c}
                    for c in px.colors.named_colorscales()
                ])
        ]),
    ])
])
# Callback function that automatically updates the first-graph based on chosen colorscale
@app.callback(
    # output 
    Output('first-graph', 'figure'),
    # input
    [Input("colorscale-dropdown", "value")]
)
# defining the function
def update_figure(colorscale):
    # ploting the scatting plot for the mean and max values
    return px.scatter(
        df2, x="mean", y="max", 
        color_continuous_scale=colorscale,
        render_mode="webgl", title="values")
#running the app server.
if __name__ == '__main__':
    app.run_server(port=4051)
```
